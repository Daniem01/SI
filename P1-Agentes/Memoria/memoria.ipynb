Descripción del REAS del agente

-R(endimiento): El objetivo es que el agente destruya la Command Center o al jugador y luego ir a la salida sin morir. El agente se considerará mejor que otro si es capaz de ganar en un tiempo menor que otro. Tambien se tendrá en cuenta la capacidad de supervivencia del agente frente a las acciones del jugador.

-E(ntorno): El entorno es un campo de batalla de 26x26 dando un total de 676 casillas. 
    ·Elementos: El entorno cuenta con muros, unos que se pueden destruir disparando y otros que no. También estarán el tanque del jugador y el tanque del agente. Ambos dispararán proyectiles.
    ·Elementos especiales: Tambien hay agua en el terreno, la cual impide el paso de los tanques por ella.

-A(ctuadores): El agente puede realizar dos acciones principales: Moverse y disparar. 
    ·Movimientos básicos: NOTHING = 0, MOVE_UP = 1, MOVE_DOWN = 2, MOVE_RIGHT = 3, MOVE_LEFT = 4.
    ·Restricciones en el movimiento: El agente no podrá moverse en diagonal, si lo intenta el agente no se moverá.
    ·Disparar: El agente podrá accionar un disparo si lo cree preciso.
    ·Acciones combinadas: El agente podrá moverse y disparar a la vez.

-S(ensores): El sensor principal del agente es un array llamado perception en cual almacena la información del entorno.
    ·(0-3): Le dice al agente lo que tiene al lado arriba, abajo, izquierda y derecha y que es basado en un código numérico.
    ·(4-7): Le dice al agente a la distancia que se encuentran los anteriores objetos detectados.
    ·(8-9): Le dice al agente las cordenadas del enemigo.
    ·(10-11): Le dice la posición de la Command Center que debe atacar.
    ·(12-13): La posición actual del agente. 
    ·(14): Le dice al agente si puede disparar.
    ·(15): Le dice al agente que salud tiene, cuantos puntos de vida le quedan.
    ·(16-17): Posición de los potenciadores.
    ·(18-19): Posición de la salida.
    ·(20): Tiempo trancurrido en la partida.

    Cualquier posición que tenga valores negativos significará que ya no se encuentra disponible en el mapa.

Clasificación del Entorno
-Observación: Es totalmente observable ya que el tanque recibe toda la información del mapa además del array perception. No hay ningún tipo de "niebla de guerra".

-Determinismo: Es estocástico ya que hay factores en el tablero que aparecen de forma randomizado y cambian la forma de ver el juego estrategicamente hablando. Además las acciones del agente random no se pueden predecir.

-Secuencial: Es secuencial ya que lo que pase en un turno afecta al siguiente. Por ejemplo si rompes en este turno un muro para los siguientes turnos puedes haber creado un camino nuevo.

-Dinámico: Es dinámico ya que el entorno va cambiando a la vez que el agente lo procesa.

-Híbrido: Es discreto en cuanto a la creación y distribución del mapa pero el movimiento es continuo ya que el tanque no salta de casilla a casilla, el enunciado dice que puede estar entre dos de ellas.

-Multiagente: Es multiagente ya que el agente tiene que interactuar con otros agentes como el jugador.

Tipo de Agente
Nuestro agente sera un Agente Reactivo Simple ya que reaccionará y hará una acción en base al entorno basado en la maquina de estados por la que se rige.
